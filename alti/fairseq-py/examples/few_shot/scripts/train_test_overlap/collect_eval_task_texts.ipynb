```python
import glob
import re
import json
import pandas as pd
```

```python
records = []

# The files below were generated using the following command
# python examples/few_shot/scripts/experiments/schedule_jobs_few_shot.py -t intermediate_eval_v1 -m 125M_gpt3_setting -o /checkpoint/danielsimig/train_test_dedupe/eval_outputs --nshot 0  --slurm-partition learnaccel

for path in glob.glob("/checkpoint/danielsimig/train_test_dedupe/eval_outputs/*/*fs0_seed0_predictions.jsonl"):
    task = re.match(".*task\.(.*)_tmp.*", path).groups()[0]
    
    with open(path) as f:
        for i, line in enumerate(f):
            line_dict = json.loads(line)
            
            id = line_dict['id']
            id = i if id == 'unk' else id
            
            for cand in line_dict['candidates']:
                prompt = cand[1]['meta']['prompt']
                records.append(
                    {
                        "task": task,
                        "id": id,
                        "prompt": prompt,
                    }
                )
    
    print(task)

data = pd.DataFrame(records)
```

```python
data.to_csv("/checkpoint/danielsimig/train_test_dedupe/prompts_v2.tsv", sep="\t", index=None, header=False)
```
